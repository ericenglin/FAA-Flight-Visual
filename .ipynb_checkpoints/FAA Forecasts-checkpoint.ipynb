```python
## Want to pull out 20 year rates using 1998, 2018, and 2038 data
## Filter by year then join tables together


faa_2018 = faa[faa["SYSYEAR"]==2018]
faa_2018 = faa_2018[['LOC_ID','Region', 'APORT_NAME',
                    'CITY', 'STATE', 'HUB_SIZE', 'AC',
                     'COMMUTER', 'T_ENPL']]
faa_2018.columns = ['LOC_ID','Region', 'APORT_NAME',
                    'CITY', 'STATE', 'HUB_SIZE', 'AC_2018',
                     'COMMUTER_2018', 'T_ENPL_2018']

faa_1998 = faa[faa["SYSYEAR"]==1998]
faa_1998 = faa_1998[['LOC_ID', 'AC',
                     'COMMUTER', 'T_ENPL']]
faa_1998.columns = ['LOC_ID', 'AC_1998',
                     'COMMUTER_1998', 'T_ENPL_1998']


faa_2038 = faa[faa["SYSYEAR"]==2038]
faa_2038 = faa_2038[['LOC_ID', 'AC',
                     'COMMUTER', 'T_ENPL']]

faa_2038.columns = ['LOC_ID', 'AC_2038',
                     'COMMUTER_2038', 'T_ENPL_2038']


faa_change = faa_2018.merge(faa_2038, on="LOC_ID")
faa_change = faa_change.merge(faa_1998, on="LOC_ID")

## Calculate Historic & Forecasted Growth rates
faa_change["Historic"]=faa_change["T_ENPL_2018"]-faa_change["T_ENPL_1998"]
faa_change["Historic_Perc"]=((faa_change["T_ENPL_2018"]-
                              faa_change["T_ENPL_1998"])
                                /faa_change["T_ENPL_1998"]*100)

faa_change["Future"]=faa_change["T_ENPL_2038"]-faa_change["T_ENPL_2018"]
faa_change["Future_Perc"]=((faa_change["T_ENPL_2038"]-
                            faa_change["T_ENPL_2018"])/
                           faa_change["T_ENPL_2018"]*100)
faa_change.describe()
```

```python
#Check to see greatest growth rate airports
faa_change.sort_values(by=["Historic"], ascending = False).head()
```

```python
#download airport data for mapping
path2 = "Airports.xlsx"
airport = pd.read_excel(path2, sheetname='Sheet1')
airport["LOC_ID"]=airport["LOCID"].str.strip()
faa_change["LOC_ID"]=faa_change["LOC_ID"].str.strip()
faa_change2 = faa_change.merge(airport, on="LOC_ID")
faa_change3 = faa_change2
```

```python
#recalculate lat long coordinates to make usable form
lat = []
for x in faa_change3["LATITUDE"]:
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1, 
                 "1":0, "0":0, "6":0,"2":0,"8":0,"5":0,"9":0, "7":0, "4":0, "3":0}
    x=str(x)
    if x=="0":
        x="01"
    new_dir = x[-1:]
    x=x[:-1]
    x=x.replace("-", "")
    x=x.replace(".", "")
    x=float(x)
    y=direction[new_dir]*x/36000000
    lat.append(y)
```

```python
long = []
for x in faa_change3["LONGITUDE"]:
    direction = {'N':-1, 'S':1, 'E': -1, 'W':1, "1":0, "0":0, 
                 "1":0, "0":0, "6":0,"2":0,"8":0,"5":0,"9":0, "7":0, "4":0, "3":0}
    x=str(x)
    if x=="0":
        x="01"
    new_dir = x[-1:]
    x=x[:-1]
    x=x.replace("-","")
    x=float(x)
    y=direction[new_dir]*x/-3600
    long.append(y)
```

```python
faa_change2["LONGITUDE"]=long
faa_change2["LATITUDE"]=lat
```

```python
long2 = []

for x in faa_change2["LONGITUDE"]:
    y = str(x)
    long2.append(y)

lat2 = []
for x in faa_change2["LATITUDE"]:
    y = str(x)
    lat2.append(y)

faa_change2["long"]=long2
faa_change2["lat"]=lat2
```

```python
faa_change3=faa_change2.loc[faa_change2['LONGITUDE'] <-50]
faa_change3=faa_change3.loc[faa_change3['LATITUDE'] >0]


faa_change3.plot(kind="scatter", x="LONGITUDE", y="LATITUDE", alpha=0.4)
plt.show()
```

```python
faa_change2['Coords']= "(" + faa_change2['lat'] + ", " + faa_change2['long'] + ")"
faa_change2[['Coords','lat','long','LATITUDE','LONGITUDE']].head()
```

```python
def merc(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)
```

```python
faa_change2=faa_change2.loc[faa_change2['LONGITUDE'] <-50]
faa_change2=faa_change2.loc[faa_change2['LATITUDE'] >1]
```

```python
faa_change2['coords_x'] = faa_change2['Coords'].apply(lambda x: merc(x)[0])
faa_change2['coords_y'] = faa_change2['Coords'].apply(lambda x: merc(x)[1])
```

```python
from bokeh.plotting import figure, show, output_notebook, save, output_file
from bokeh.tile_providers import CARTODBPOSITRON
from bokeh.models import ColumnDataSource, HoverTool

faa_change2["T_ENPL_2018_adjusted"]=faa_change2['T_ENPL_2018']/1000000

source = ColumnDataSource(data=dict(
                        x=list(faa_change2['coords_x']), 
                        y=list(faa_change2['coords_y']),
                        enplanements=list(faa_change2['T_ENPL_2018']),
                        sizes=list(faa_change2['T_ENPL_2018_adjusted']),
                        oldgrowth=list(faa_change2['Historic_Perc']),
                        newgrowth=list(faa_change2['Future_Perc']),
                        airportname=list(faa_change2['APORT_NAME_x']),
                        city=list(faa_change2['CITY_x'])))


hover = HoverTool(tooltips=[
    ("airport", "@airportname"),
    ("city", "@city"),
    ("enplanements","@enplanements{0,0}"),
    ("Forecasted 20-Year Growth Rate","@newgrowth{0.1f}%"),
    ("Historical 20-Year Growth Rate","@oldgrowth{0.1f}%")
])


p = figure(plot_width=1000, plot_height=700,
           x_range=(-14000000, -7500000), y_range=(2800000, 6000000),
           x_axis_type="mercator", y_axis_type="mercator",
           tools=[hover,'reset', 'box_zoom', 'wheel_zoom','save'],
          title="2018 FAA Enplanements")

p.add_tile(CARTODBPOSITRON)
p.title.text_font_size = '14pt'


p.circle(x='x',
         y='y',
         source=source,
         size='sizes',
         line_color="#FF0000", 
         fill_color="#FF0000",
         fill_alpha=0.05)

output_file("FAA Enplanements 2018.html")

output_notebook()
save(p)
```

```python
#download airport data to reconfigure the hub LOCID's, cleaned manually
path3 = "2038 FAA Forecasts Slim Revised.xlsx"
faa_slim = pd.read_excel(path3, sheetname='Sheet 1')
```

```python
faa_slim=faa_slim.drop(['LOCID', 'REGION',
       'APORT_NAME_y', 'CITY_y', 'STATE_y', 'SITE', 'NASPNO','FAC_CLASS', 'FAC_LEVEL', 'ATCT_FLAG', 'PUBLISHED',
       'SER_VOL', 'HUB_SIZE_y', 'RELIVER', 'RE_ARPT_ID',
       'CENTER_ID', 'IOPS_ID', 'TR_PRIMARY', 'OPS_FLAG', 'IOPS_NOADD',
       'REG_INPUT', 'REG_SYEAR', 'REG_EYEAR', 'OEP35'], axis=1)
```

```python
hubs = faa_slim.loc[faa_slim.HUB_LOCID.notnull()]
hubs = hubs.sort_values(by="T_ENPL_2018", ascending=False)
```

```python
hubs_sum = hubs.groupby(by="HUB_LOCID").agg(
    {
        'coords_x':'first', 'coords_y':'first',
        'T_ENPL_2018':'sum',  'T_ENPL_1998':'sum',
         'T_ENPL_2038':'sum', 'CITY_x':'first', 'STATE_x':'first'
    })
```

```python
hubs_sum['HUB']=hubs_sum.index
hubs_sum.head()
```

```python
## Now That Grouped, Need To Re-Calculate Historic & Forecasted Growth rates
hubs_sum["Historic"]=hubs_sum["T_ENPL_2018"]-hubs_sum["T_ENPL_1998"]
hubs_sum["Historic_Perc"]=((hubs_sum["T_ENPL_2018"]-
                              hubs_sum["T_ENPL_1998"])
                                /hubs_sum["T_ENPL_1998"]*100)

hubs_sum["Future"]=hubs_sum["T_ENPL_2038"]-hubs_sum["T_ENPL_2018"]
hubs_sum["Future_Perc"]=((hubs_sum["T_ENPL_2038"]-
                            hubs_sum["T_ENPL_2018"])/
                           hubs_sum["T_ENPL_2018"]*100)
hubs_sum.describe()
```

```python
listt=[]
for x in hubs_sum["Future_Perc"]:
    y = float(x)
    listt.append(y)

hubs_sum["Future_Perc"]=listt
```

```python
hubs_sum = hubs_sum.loc[hubs_sum.Future_Perc.notnull()]
len(hubs_sum)
```

```python
hubs_sum['Historic_Perc2']=hubs_sum['Historic_Perc']/100
hubs_sum['Future_Perc2']=hubs_sum['Future_Perc']/100
```

```python
hub_status=[]
for x in hubs_sum['T_ENPL_2018']:
    n=x/840817923
    if n>0.01:
        hub_status.append('Large Hub')
    elif n>0.0025:
        hub_status.append('Medium Hub')
    elif n>0.0005:
        hub_status.append('Small Hub')
    else:
        hub_status.append('Nonhub')
```

```python
hubs_sum["Hub_Status"]=hub_status
```

```python
colormap = {'Large Hub': 'red', 'Medium Hub': 'green', 'Small Hub': 'blue','Nonhub':'purple'}
#colormap
colors = [colormap[x] for x in hubs_sum["Hub_Status"]]
hubs_sum['colors']=colors
```
